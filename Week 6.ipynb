# Question 1 — Temperature Logger 

```python
import math

station_name="Kathmandu Weather Station"
temperatures = [18.4, 22.1, 15.7, 29.3, 11.8, 25.6, 19.2]

def get_average_temperature(temps):
    if not temps:
        return 0
    return sum(temps) / len(temps)

def get_deviation(temps):
    mean = get_average_temperature(temps)
    variance = sum((temp - mean) ** 2 for temp in temps) / len(temps)
    deviation = math.sqrt(variance)
    return deviation

def get_summary(temps):
    print(f"Temperature Summary for {station_name}:")
    print("Minimum Temperature:", min(temps), "°C")
    print("Maximum Temperature:", max(temps), "°C")
    print("Average Temperature:", round(get_average_temperature(temps), 2), "°C")
    print("Standard Deviation:",round(get_deviation(temps), 2), "°C")

get_summary(temperatures)
```

# Question 2 — Bill Splitter

```python
import random

random.seed(42)

friends = ["Ramesh", "Sunita", "Bikash", "Anjali", "Dipak"]
total_bill = 3750

def split_bill_among_friends(friends, total_bill):
    return total_bill / len(friends)

def pick_lucky_friend(friends):
    return random.choice(friends)

def final_summary(friends, total_bill):
    share = split_bill_among_friends(friends, total_bill)
    lucky_friend = pick_lucky_friend(friends)
    print("Bill Split Summary:")
    print("-"*30)

    for friend in friends:
        print(f"{friend} pays: {round(share, 2)} NPR")
    
    lucky_total= share + 50

    print("-"*30)

    print(f"Lucky Person: {lucky_friend}")
    print(f"{lucky_friend} has to pay extra NRP 50")
    print(f"Total Bill for {lucky_friend}: NPR {lucky_total:.2f}")

final_summary(friends, total_bill)
```

# Question 3 — Exam Scheduler

```python
import datetime

college_name = "Bhaktapur Multiple College"

start_date = "2025-05-01"

exams = [
    ("Python Programming", 0),
    ("Data Structures",    3),
    ("Database Systems",   6),
    ("Computer Networks",  10),
    ("Mathematics",        14),
]

def parse_date(date_str):
    return datetime.datetime.strptime(date_str, "%Y-%m-%d")

def get_exam_schedule(start_date, days):
    start_date = parse_date(start_date)
    exam_date=start_date + datetime.timedelta(days=days)
    return exam_date.strftime("%Y-%m-%d")
def print_schedule(start_str,exams):
    print(college_name)
    print("Exam Schedule:")
    for subject, day in exams:
        print(f"{subject:<20}: {get_exam_schedule(start_str, day)}")

print_schedule(start_date, exams)
```
