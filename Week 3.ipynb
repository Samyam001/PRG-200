# Question 1 - ATM Withdrawal Validator

```python
balance=int(input("Enter current account balance: "))
daily_withdrawn=int(input("Enter the amount already withdrawn today: "))
amount=int(input("Enter the amount You want to withdraw: "))

daily_limit=50000

if amount %500 !=0:
    print("Withdrawn amount must be multiple of 500")
elif amount>balance:
    print("Insufficent balance")
elif daily_withdrawn+amount>daily_limit:
    print(" Daily withdrawal limit reached.")
else:
    balance-=amount
    daily_withdrawn+=amount
    print("Withdrawal Sucessfull")
    print(f"Your daily withdrawal after the transaction: {daily_withdrawn}")
    print(f"Your current balance after withdrawal: {balance}")
    
```

# Question 3 - Inventory Restock Alert

```python
inventory = [
    {"item": "Rice", "stock": 5, "threshold": 10},
    {"item": "Eggs", "stock": 24, "threshold": 12},
    {"item": "Milk", "stock": 3, "threshold": 6},
    {"item": "Bread", "stock": 8, "threshold": 5},
    {"item": "Chicken", "stock": 0, "threshold": 4},
    {"item": "Cooking Oil", "stock": 2, "threshold": 3},
]

restock_count = 0

for product in inventory:
    if product["stock"] < product["threshold"]:
        print(
            f"Restock Alert: {product['item']} :Left {product['stock']}")
        restock_count += 1

print(f"\nTotal items needing restock: {restock_count}")
for product in inventory:
    if product["stock"] < product["threshold"]:
        print(f"{product['item']}")
```

# Question 4- Password Strength Checker

```python
passwords = ["hello", "Hello123", "H3ll0@World", "12345678", "MyP@ss!"]
special_chars = "!@#$%^&*"

for password in passwords:
    print(f"\nChecking password: {password}")

    missing = []
    if len(password) < 8:
        missing.append("At least 8 characters long")
        
    has_upper = False
    for ch in password:
        if ch.isupper():
            has_upper = True
            break

    if not has_upper:
        missing.append("At least one uppercase letter")


    has_lower = False
    for ch in password:
        if ch.islower():
            has_lower = True
            break

    if not has_lower:
        missing.append("At least one lowercase letter")

    has_digit = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
            break

    if not has_digit:
        missing.append("At least one digit")

    has_special = False
    for ch in password:
        if ch in special_chars:
            has_special = True
            break

    if not has_special:
        missing.append("At least one special character (!@#$%^&*)")

    if len(missing) == 0:
        print("Strong Password")
    else:
        print("Weak Password")
        print("Missing Criteria:")
        for item in missing:
            print("-", item)
```

# Question 5-Taxi Fare

```python
trips = [
    {"distance": 1.5, "hour": 14},
    {"distance": 5.0, "hour": 22},
    {"distance": 12.0, "hour": 3},
    {"distance": 8.5, "hour": 10},
    {"distance": 2.0, "hour": 23},
]

for trip in trips:

    distance = trip["distance"]
    hour = trip["hour"]
    fare = 150

    if distance > 2:
        if distance <= 10:
            fare += (distance - 2) * 35
        else:
            fare += 8 * 35
            fare += (distance - 10) * 28
    if hour >= 22 or hour < 5:
        fare = fare + (fare * 0.10)

    print("Distance:", distance, "km")
    print("Travel Hour:", hour)
    print("Total Fare: NPR", round(fare, 2))
    print("-" * 30)
```
