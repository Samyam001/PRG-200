# Question 1 — Bank Account Manager

```python
class BankAccount():
    def __init__(self,name,account_number,balance=0):
        self.name=name
        self.account_number=account_number
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"{self.name}: Deposited {amount}. New Balance: {self.balance}")
    def withdraw(self,amount):
        if amount >self.balance:
            print(f"{self.name}: Insufficent Balance")
        else:
            self.balance-=amount
            print(f"{self.name}: Withdrawn {amount}. New Balance: {self.balance}")
    def get_balance(self):
        print(f"Name:{self.name}: Balance:{self.balance}")
        

accounts = [("Ramesh Thapa", "A001", 5000),
("Sunita Karki", "A002", 0),
("Bikash Rai", "A003", 12000),]

bank_account={}

for name, acc_no, balance in accounts:
    bank_account[acc_no]=BankAccount(name,acc_no,balance)

bank_account["A002"].deposit(3000)
bank_account["A003"].withdraw(15000)
bank_account["A001"].withdraw(2000)

print("New Balance")
for account in bank_account.values():
    account.get_balance()
```

# Question 2 — Student Report Card

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def average(self):
        return sum(self.marks)/len(self.marks)
    
    def grades(self):
        avg=self.average()

        if avg >= 80:
            return "A"
        elif avg>=65:
            return "B"
        elif avg>=50:
            return "C"
        elif avg>=40:
            return "D"
        else:
            return "F"
        
    def display(self):
        avg=self.average()
        grades=self.grades()

        if avg>=40:
            result = "Pass"
        else:
            result= "Fail"

        print("Name:", self.name)
        print("Average:", round(avg, 2))
        print("Grade:", grades)
        print("Result:", result)
        print("-" * 25)

    
students = [ ("Aarav", [78, 85, 60, 90, 72]),
    ("Sita", [45, 50, 38, 60, 55]),
    ("Bishal", [30, 25, 40, 35, 28]),
    ("Priya", [90, 88, 95, 92, 87]),]

for name, marks in students:
    student=Student(name,marks)
    student.display()
```

# Question 3 — Food Delivery App

```python
class DeliveryPartner:
    def __init__(self, name, partner_id, deliveries):
        self.name = name
        self.partner_id = partner_id
        self.deliveries = deliveries
    def total_earnings(self):
        pass
    def display(self):
        print(f"Name: {self.name}")
        print(f"Deliveries: {self.deliveries}")
        print("Total Earnings: RS", self.total_earnings())
        print("---------------------------------------")


class BikeRider(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, km_travelled):
        super().__init__(name, partner_id, deliveries)
        self.km_travelled = km_travelled
    def total_earnings(self):
        return (80 * self.deliveries) + (5 * self.km_travelled)


class Walker(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, rainy_deliveries):
        super().__init__(name, partner_id, deliveries)
        self.rainy_deliveries = rainy_deliveries
    def total_earnings(self):
        return (60 * self.deliveries) + (50 * self.rainy_deliveries)


class CarDriver(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, fuel_cost):
        super().__init__(name, partner_id, deliveries)
        self.fuel_cost = fuel_cost
    def total_earnings(self):
        return (120 * self.deliveries) - self.fuel_cost

partners = [
    BikeRider("Santosh Rai", "B-01", 15, 42),
    Walker("Kabita Maharjan", "W-01", 18, 5),
    CarDriver("Roshan KC", "C-01", 20, 850),
]

for partner in partners:
    partner.display()

highest = partners[0]

for partner in partners:
    if partner.total_earnings() > highest.total_earnings():
        highest = partner

print("Highest Earning Partner:")
print(highest.name, ":", "NPR", highest.total_earnings())
```

# Question 4 — Bus Ticket Booking

```python
class Bus:
    def __init__(self, route, total_seats):
        self.route = route
        self.total_seats = total_seats
        self.booked = {} 
        self.dup_booking=[]

    def book_seat(self, seat_number, passenger_name):
        if seat_number in self.booked:
            print(f"Seat {seat_number} already booked")
            self.dup_booking.append(passenger_name)
        else:
            self.booked[seat_number] = passenger_name

    def available_seats(self):
        return self.total_seats - len(self.booked)

    def passenger_list(self):
        print("\nPassenger List:")
        for seat, passenger in sorted(self.booked.items()):
            print(f"Seat {seat}: {passenger}")
    
    def duplicate_booking(self):
        for passenger in self.dup_booking:
            print(f" * {passenger} ")


bus = Bus("Kathmandu - Pokhara", 10)

bookings = [
    (3, "Ramila Shrestha"),
    (7, "Deepak Gurung"),
    (3, "Anita Rai"),     
    (1, "Prakash Magar"),
    (7, "Suman Tamang"),  
]

for seat, name in bookings:
    bus.book_seat(seat, name)

print("\nAvailable seats:", bus.available_seats())
bus.passenger_list()
print("\nDuplicate Seat")
bus.duplicate_booking()
```
