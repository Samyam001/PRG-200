```python
inventory = {
"rice": {"price": 120, "stock": 20},
"milk": {"price": 90, "stock": 10},
"bread": {"price": 60, "stock": 15},
"eggs": {"price": 15, "stock": 30}
}

cart = {
"rice": 2,
"milk": 3,
"eggs": 12
}

def process_order(inventory, cart):
    total_cost=0
    for item,quantity in cart.items():
        if item in inventory:
            if inventory[item]["stock"] >= quantity:
                cost = inventory[item]["price"] * quantity
                total_cost += cost
                inventory[item]["stock"] -= quantity
            else:
                print(f"Not enough stock for {item}. Available: {inventory[item]['stock']}")
        else:
            print(f"{item} is not available in the inventory.")
    return total_cost


def remaining_stock(inventory):
    print("\nRemaining Stock:")
    for item, details in inventory.items():
        print(f"{item.capitalize()}: {details['stock']} units")


print("----Bill----")
for item, quantity in cart.items():
    if item in inventory:
        cost = inventory[item]["price"] * quantity
        print(f"{item.capitalize()}: {quantity} x {inventory[item]['price']} = Rs {cost}")

total_cost = process_order(inventory, cart)
print(f"Total cost: Rs {total_cost}")
remaining_stock(inventory)

```

```python
bs_months = ["Baisakh","Jestha","Ashadh","Shrawan","Bhadra","Ashwin",
"Kartik","Mangsir","Poush","Magh","Falgun","Chaitra"]
customers = [
{"name": "Ramesh Thapa", "date": "1985-06-24", "cal": "AD", "need": "BS", "style":
"full"},
{"name": "Sunita Karki", "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
{"name": "Bikash Rai", "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
{"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style":
"full"}, ]

def convert_date(date_str, from_cal, to_cal):
    if from_cal == "AD" and to_cal == "BS":
        year, month, day = map(int, date_str.split("-"))
        bs_year = year + 56
        bs_month = (month + 8) % 12
        if bs_month == 0:
            bs_month = 12
        bs_day = day
        return f"{bs_year}-{bs_month:02d}-{bs_day:02d}"
    elif from_cal == "BS" and to_cal == "AD":
        year, month, day = map(int, date_str.split("-"))
        ad_year = year - 56
        ad_month = (month - 8) % 12
        if ad_month <= 0:
            ad_month += 12
        ad_day = day
        return f"{ad_year}-{ad_month:02d}-{ad_day:02d}"
    else:
        return "Invalid calendar conversion"
    
print("\n----Date Conversions----")
for customer in customers:
    converted_date = convert_date(customer["date"], customer["cal"], customer["need"])
    print(f"{customer['name']}: {customer['date']} ({customer['cal']}) -> {converted_date} ({customer['need']})")
```

```python
text = """
Nepal is a beautiful country. Nepal has Mount Everest.
Everest is the highest mountain in the world. Many tourists
visit Nepal every year to see Everest and other mountains.
Nepal is known for its mountains and natural beauty.
"""

def word_count(text):
    words = text.split()
    count = {}

    for word in words:
        word = word.strip(".,").lower()

        if word in count:
            count[word] += 1
        else:
            count[word] = 1

    return count

print("---- Word Count ----")
counts = word_count(text)

items = list(counts.items())

for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]

top_three = items[:3]

print("\nTop 3 most common words:")
for word, count in top_three:
    print(f"{word}: {count} times")
```

```python
accounts = {
"A001": {"name": "Ramesh Thapa", "balance": 15000, "pin":
"1234"},
"A002": {"name": "Sunita Karki", "balance": 8500, "pin":
"5678"},
"A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
}

def atm(account_id,pin,action,amount=0):
    if account_id in accounts:
        account = accounts[account_id]
        if pin == account["pin"]:
            if action == "withdraw":
                if amount <= account["balance"]:
                    account["balance"] -= amount
                    return f"Withdrawal successful. New balance: {account['balance']}"
                else:
                    return "Insufficient funds."
            elif action == "deposit":
                account["balance"] += amount
                return f"Deposit successful. New balance: {account['balance']}"
            elif action == "check_balance":
                return f"Current balance: {account['balance']}"
        else:
            return "Invalid PIN."
    else:
        return "Account not found."
print("\n----ATM Simulation----")
print(f"('A001', '1234', 'check_balance'): {atm('A001', '1234', 'check_balance')}")
print(f"('A002', '0000', 'withdraw', 2000): {atm('A002', '0000', 'withdraw', 2000)}")
print(f"('A003', '9012', 'deposit', 3000): {atm('A003', '9012', 'deposit', 3000)}")
print(f"('A004', '1111', 'check_balance'): {atm('A004', '1111', 'check_balance')}")
```
